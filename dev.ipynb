```python
import numpy as np

ANALYSIS_TREND_COLUMNS = ['開始時刻', '分類', '高値', '安値']
#検証する左右の期間数
PERIOD = 5
```

```python
analysis_trend_df = []
test = [0,'スイングハイ',0,None]

analysis_trend_df.append(test)

test = [0,'スイングロウ',None,0]
analysis_trend_df.append(test)

analysis_trend_df = pd.DataFrame(data=np.array(analysis_trend_df), columns=ANALYSIS_TREND_COLUMNS)

print(analysis_trend_df)
```

```python
#高値か安値かの検証をする値
predicted_value = one_BTC.df.iloc[41]
#左右各5期間分のデータ
left_redicted_values = one_BTC.df.iloc[36 : 41]
right_redicted_values = one_BTC.df.iloc[95 : 100]

#検証する値が左右のデータのどの値よりも大きい場合はスイングハイ、小さい場合はスイングロウ
#どちらでもない場合は飛ばす
for i in range(0, PERIOD):
    #左右のいずれかの高値が検証したい期間の高値より大きい場合は検証値はスイングハイではないと
    #判断してスイングロウになり得るか検証するためにbreakする
    if predicted_value['高値'] < left_redicted_values.iloc[i]['高値'] or predicted_value['高値'] < right_redicted_values.iloc[i]['高値']:
        print(f'検証値: {predicted_value['高値']} | 左: {left_redicted_values.iloc[i]['高値']} | 右: {right_redicted_values.iloc[i]['高値']}')
        break
else:
    analysis_trend_df = pd.DataFrame(
                            data=np.array([[
                                predicted_value['開始時刻'],
                                'スイングハイ',
                                predicted_value['高値'],
                                None
                            ]]),
                            columns=ANALYSIS_TREND_COLUMNS
                        )
    print(analysis_trend_df)
    
```

```python
analysis_trend_array = []

for i in range(0, 100):
    #高値か安値かの検証をする値
    predicted_value = one_BTC.df.iloc[i]
    #左右各5期間分のデータ
    #左右どちらかに5期間分のデータがない場合は空のDataframeを入れて片方の期間と比べる
    left_redicted_values = one_BTC.df.iloc[
        i - 5 : i
    ] if i >= 5 else pd.DataFrame()
    
    right_redicted_values = one_BTC.df.iloc[
        i + 1 : i + 6
    ] if i + 5 <= 100 else pd.DataFrame()

    # #検証する値が左右のデータのどの値よりも大きい場合はスイングハイ、小さい場合はスイングロウ
    # #どちらでもない場合は飛ばす
    for n in range(0, 5):
        #左右のいずれかの高値が検証したい期間の高値より大きい場合は検証値はスイングハイではないと
        #判断してスイングロウになり得るか検証するためにbreakする
        is_left_valid = False if left_redicted_values.empty else predicted_value[
            '高値'] < left_redicted_values.iloc[n]['高値']
        
        is_right_valid = False if right_redicted_values.empty else predicted_value[
            '高値'] < right_redicted_values.iloc[n]['高値']

        if is_left_valid or is_right_valid:
            break
    else:
        swing_high = [
                        predicted_value['開始時刻'],
                        'スイングハイ',
                        predicted_value['高値'],
                        None
                    ]
        analysis_trend_array.append(swing_high)
```

```python
target_data = analysis_one_hour_BTC.df.iloc[17]['高値']
latest_data = analysis_one_hour_BTC.df.iloc[16]

#validate_down_trend(現在のローソク足, 前回データ)
conversion_value = analysis_one_hour_BTC.df.iloc[latest_data.name - 1]['高値']
current_target_value = analysis_one_hour_BTC.df.iloc[latest_data.name - 2]['安値']

#現在のトレンドが下降の時に実行されて一つ前のローソク足の分類で処理を分岐させる

#前回データがスイングハイの場合は安値が切り下がっていることを確認する
if latest_data['分類'] == 'スイングハイ':

    is_low_price_down = current_target_value > target_data

    #安値が前回安値を下回っている場合は下降トレンドの継続と判断して、
    #トレンドに'下降'を入れて転換値を前回高値、直近目標値に現在のローソク足データ
    #を入れる
    if is_low_price_down:
        conversion_value = latest_data['高値']
        current_target_value = target_data
    
    #安値の更新に失敗した場合でも高値は転換値を超えていないためトレンドは継続と判断して、
    #トレンドに'下降',転換値と直近目標値を前のデータのカラムから取得する
    
else:
    is_high_price_down = conversion_value > target_data

    #転換値を現在のローソク足データに更新して返す
    if is_high_price_down:
        conversion_value = target_data
        current_target_value = latest_data['安値']
    
    #転換値を超えた場合はトレンドの転換と判断して、トレンド,転換値,直近目標値すべてをNaNで返す
    else:
        trend = np.nan
        conversion_value = np.nan
        current_target_value = np.nan

trend, conversion_value, current_target_value
```
